## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Read data

```python
input_path = '../../DATA/Commuting-Distance-by-Personal-Characteristics/generated_data/'
input_file = 'transformed_data.xlsx'
```

```python
df = pd.read_excel(input_path + input_file, usecols=[1, 2, 3], sheet_name='age')
df
```

The generated data contains the information about Australian's commuting distance by personal characteristics in 2016. Column 'Age group (AGEP)' includes different age groups, while 'Average commuting distance (kilometres)' column shows the average commuting distance for these age groups. The average commuting distance in Australia is 15.98km where people aged 30 to 39 have the highest average commuting distance (16.59km).   

## Visualize data

Here, we'd like to visualize the generated data. Below code demonstrates how to visualize the data by using Matplot library.

**Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = list(df['Age group (AGEP)'])[:-1]
AVG_distances = list(df['Average commuting distance \n(kilometres)'])[:-1]

x_pos = [i for i, _ in enumerate(x)]
plt.figure(figsize=(22,14))
plt.bar(x_pos, AVG_distances, color='#F86051')
plt.xlabel("Age groups", fontsize=22)
plt.ylabel("Kilometres", fontsize=22)
plt.title("Average commuting distance (km) by age", fontsize=28)
plt.ylim((0, 25))   # set the ylim to bottom, top
for index, value in enumerate(AVG_distances):
    plt.text(index - 0.22, value - 0.8, str(round(value, 2)), fontsize=18)

plt.xticks(x_pos, x, fontsize=18, rotation = 90)

plt.savefig('img/Commuting distance by age.pdf', bbox_inches='tight')
plt.show()
```

People aged between 20 - 69 years report a similar average daily commuting distance of ( between 16.56 and 16.59km). While younger people aged  between 15-19 years  have the lowest average commuting distance (11.3km), followed by people aged between 70 -79 years (13.18km)

**People aged between 30 - 39 years report 16.59km of average daily commuting distance in Australia in 2016.**
